```python
import cv2
import os
import matplotlib.pyplot as plt
from skimage.metrics import structural_similarity as ssim
from skimage.metrics import peak_signal_noise_ratio as psnr
from skimage.transform import resize
import numpy as np

# Define the mapping of super-resolution image filenames to original image filenames
pairs = {
    "C1_thinF_IMG_20150604_104722_cell_9_RCAN_BIX4-official.png": "C1_thinF_IMG_20150604_104722_cell_9.png",
    "C214ThinF_IMG_20151106_131748_cell_148_RCAN_BIX4-official.png": "C214ThinF_IMG_20151106_131748_cell_148.png",
    "C226ThinF_IMG_20151112_131408_cell_102_RCAN_BIX4-official.png": "C226ThinF_IMG_20151112_131408_cell_102.png",
    "C221ThinF_IMG_20151106_155857_cell_97_RCAN_BIX4-official.png": "C221ThinF_IMG_20151106_155857_cell_97.png"
}

# Paths to the folders containing the original and super-resolution images
original_dataset_path = r'E:\SRClass\archive\cell_images\Uninfected'
super_res_dataset_path = r'D:\dataset\Uninfected'

# Function to load images
def load_image_pair(super_res_filename, original_filename):
    sr_image = cv2.imread(os.path.join(super_res_dataset_path, super_res_filename))
    original_image = cv2.imread(os.path.join(original_dataset_path, original_filename))
    return sr_image, original_image

# Function to show image comparison with axes and metrics
def show_image_comparison(image_pairs):
    n = len(image_pairs)
    plt.figure(figsize=(16, 8))
    for i, (sr_img, original_img) in enumerate(image_pairs, start=1):
        # Check if original and super-resolution images have the same dimensions
        if sr_img.shape != original_img.shape:
            # Resize the original image to match the super-resolution image dimensions
            original_img = cv2.resize(original_img, (sr_img.shape[1], sr_img.shape[0]), interpolation=cv2.INTER_CUBIC)

        # Resize super-resolution image to be 2x the size of the original image
        sr_img_resized = cv2.resize(sr_img, (original_img.shape[1]*2, original_img.shape[0]*2), interpolation=cv2.INTER_CUBIC)
        sr_img_gray = cv2.cvtColor(sr_img_resized, cv2.COLOR_BGR2GRAY)
        original_img_gray = cv2.cvtColor(original_img, cv2.COLOR_BGR2GRAY)

        ssim_value = ssim(original_img_gray, sr_img_gray)
        psnr_value = psnr(original_img_gray, sr_img_gray)

        # Show the original image
        ax1 = plt.subplot(2, n, i)
        ax1.imshow(cv2.cvtColor(original_img, cv2.COLOR_BGR2RGB))
        ax1.set_title('Original')
        ax1.axis('on')  # Show axes

        # Show the super-resolution image
        ax2 = plt.subplot(2, n, n + i)
        ax2.imshow(cv2.cvtColor(sr_img_resized, cv2.COLOR_BGR2RGB))
        ax2.set_title(f'Super-Resolution\nSSIM: {ssim_value:.4f}, PSNR: {psnr_value:.2f} dB')
        ax2.axis('on')  # Show axes

    plt.tight_layout()
    plt.show()

   

# Load and display all image pairs
image_pairs = []
for sr_filename, original_filename in pairs.items():
    try:
        sr_image, original_image = load_image_pair(sr_filename, original_filename)
        image_pairs.append((sr_image, original_image))
    except Exception as e:
        print(f"An error occurred while processing files: {sr_filename} and {original_filename}")
        print(e)

show_image_comparison(image_pairs)
```
